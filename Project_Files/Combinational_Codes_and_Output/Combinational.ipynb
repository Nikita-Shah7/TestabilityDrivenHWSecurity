```python
import os
import pandas as pd
import chardet

# Function to detect encoding of a file
def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        raw_data = f.read()
        result = chardet.detect(raw_data)
        encoding = result['encoding']
        confidence = result['confidence']
        return encoding, confidence

# Function to parse the circuit details from a single text file with detected encoding
def parse_circuit_details(file_path, encoding):
    circuit_details = {
        'No. of Primary Inputs': None,
        'No. of Primary Outputs': None,
        'No. of Gates': None,
        'No. of DFFs': None,
        'No. of Nodes': None,
        'Execution Time': None,
        'CC0': [],
        'CC1': [],
        'SC0': [],
        'SC1': []
    }

    with open(file_path, 'r', encoding=encoding) as file:
        lines = file.readlines()
        # Extracting primary inputs, outputs, gates, DFFs, nodes, execution time
        for line in lines:
            if line.startswith("No. of Primary Inputs:"):
                circuit_details['No. of Primary Inputs'] = int(line.split(': ')[1].strip())
            elif line.startswith("No. of Primary Outputs:"):
                circuit_details['No. of Primary Outputs'] = int(line.split(': ')[1].strip())
            elif line.startswith("No. of Gates:"):
                circuit_details['No. of Gates'] = int(line.split(': ')[1].strip())
            elif line.startswith("No. of DFFs:"):
                circuit_details['No. of DFFs'] = int(line.split(': ')[1].strip())
            elif line.startswith("No. of Nodes:"):
                circuit_details['No. of Nodes'] = int(line.split(': ')[1].strip())
            elif line.startswith("Execution Time:"):
                circuit_details['Execution Time'] = int(line.split(': ')[1].strip().split()[0])  # assuming it's in ns

            # Extracting signal data and storing values
            if line.strip().startswith("G"):
                parts = line.split()
                signal_name = parts[0]
                values = list(map(int, parts[1:5]))
                if signal_name in circuit_details:
                    circuit_details[signal_name].extend(values)

    # Update max values
    for signal in ['CC0', 'CC1', 'SC0', 'SC1']:
        if circuit_details[signal]:  # Check if list is not empty
            max_value = max(circuit_details[signal])
            circuit_details[signal] = max_value
        else:
            circuit_details[signal] = None  # Handle case where no values were found

    return circuit_details

# Function to process all files in a folder
def process_folder(input_folder, output_file):
    files = os.listdir(input_folder)
    data = {}
    for file_name in files:
        file_path = os.path.join(input_folder, file_name)
        if os.path.isfile(file_path) and file_name.endswith('.txt'):
            encoding, _ = detect_encoding(file_path)
            circuit_details = parse_circuit_details(file_path, encoding)
            data[file_name] = circuit_details

    # Create DataFrame from collected data
    df = pd.DataFrame(data)
    df = df.transpose()  # Transpose so that filenames become column headings

    # Write to Excel
    df.to_excel(output_file)

# Example usage
input_folder = '/content/s35932'  # Replace with the path to your folder containing text files
output_file = '/content/s35932/circuit_details_all.xlsx'    # Output Excel file path

process_folder(input_folder, output_file)
```

```python
import os
import pandas as pd
import chardet

# Function to detect encoding of a file
def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        raw_data = f.read()
        result = chardet.detect(raw_data)
        encoding = result['encoding']
        confidence = result['confidence']
        return encoding, confidence

# Function to parse the circuit details from a single text file with detected encoding
def parse_circuit_details(file_path, encoding):
    circuit_details = {
        'No. of Primary Inputs': 0,
        'No. of Primary Outputs': 0,
        'No. of Gates': 0,
        'No. of DFFs': 0,
        'No. of Nodes': 0,
        'Execution Time': 0,
        'Max_CC0': 0,
        'Max_CC1': 0,
        'Max_SC0': 0,
        'Max_SC1': 0
    }

    with open(file_path, 'r', encoding=encoding) as file:
        lines = file.readlines()
        # Extracting primary inputs, outputs, gates, DFFs, nodes, execution time
        for line in lines:
            if line.startswith("No. of Primary Inputs:"):
                circuit_details['No. of Primary Inputs'] = int(line.split(': ')[1].strip())
            elif line.startswith("No. of Primary Outputs:"):
                circuit_details['No. of Primary Outputs'] = int(line.split(': ')[1].strip())
            elif line.startswith("No. of Gates:"):
                circuit_details['No. of Gates'] = int(line.split(': ')[1].strip())
            elif line.startswith("No. of DFFs:"):
                circuit_details['No. of DFFs'] = int(line.split(': ')[1].strip())
            elif line.startswith("No. of Nodes:"):
                circuit_details['No. of Nodes'] = int(line.split(': ')[1].strip())
            elif line.startswith("Execution Time:"):
                circuit_details['Execution Time'] = int(line.split(': ')[1].strip().split()[0])  # assuming it's in ns

        # Extracting signal data and finding max values
        for line in lines:
            if line.strip().startswith("G_CC0") or line.strip().startswith("G_CC1") or line.strip().startswith("G_SC0") or line.strip().startswith("G_SC1"):
                parts = line.split()
                signal_name = parts[0]
                values = list(map(int, parts[1:5]))
                if signal_name.startswith("G_CC0"):
                    circuit_details['Max_CC0'] = max(circuit_details['Max_CC0'], *values)
                elif signal_name.startswith("G_CC1"):
                    circuit_details['Max_CC1'] = max(circuit_details['Max_CC1'], *values)
                elif signal_name.startswith("G_SC0"):
                    circuit_details['Max_SC0'] = max(circuit_details['Max_SC0'], *values)
                elif signal_name.startswith("G_SC1"):
                    circuit_details['Max_SC1'] = max(circuit_details['Max_SC1'], *values)

    return circuit_details

# Function to process all files in a folder
def process_folder(input_folder, output_file):
    files = os.listdir(input_folder)
    all_data = {}
    parameter_set = set()  # To store all unique parameters

    # Iterate through each file in the folder
    for file_name in files:
        file_path = os.path.join(input_folder, file_name)
        if os.path.isfile(file_path) and file_name.endswith('.txt'):
            encoding, _ = detect_encoding(file_path)
            circuit_details = parse_circuit_details(file_path, encoding)
            all_data[file_name] = circuit_details
            parameter_set.update(circuit_details.keys())  # Collect all unique parameters

    # Create a dictionary to hold the structured data
    structured_data = {param: [] for param in parameter_set}

    # Initialize the dictionary with None values
    for param in parameter_set:
        for file_name in files:
            if file_name.endswith('.txt'):
                structured_data[param].append(all_data.get(file_name, {}).get(param, None))

    # Convert to DataFrame
    df = pd.DataFrame(structured_data, index=[file_name for file_name in files if file_name.endswith('.txt')]).transpose()

    # Write to Excel
    df.to_excel(output_file)

# Example usage
input_folder = '/content/s13207'  # Replace with the path to your folder containing text files
output_file = '/content/s13207/circuit_details_all.xlsx'  # Output Excel file path

process_folder(input_folder, output_file)
```

```python
import os
import chardet
import openpyxl
import re

# Function to extract numerical values from a line
def extract_numbers(line):
    return [int(s) for s in re.findall(r'\d+', line)]

# Function to extract execution time from line
def extract_execution_time(line):
    time_str = line.split(':')[1].strip()
    # Extract numeric part
    time_value = re.search(r'\d+', time_str).group()
    return float(time_value)

# Function to process a single file
def process_file(file_path):
    # Detect encoding
    with open(file_path, 'rb') as f:
        raw_data = f.read()
        encoding_info = chardet.detect(raw_data)
        encoding = encoding_info['encoding']

    # Initialize variables to store extracted data
    parameters = {
        'Filename': os.path.basename(file_path),
        'No. of Primary Inputs': None,
        'No. of Primary Outputs': None,
        'No. of Gates': None,
        'No. of Nodes': None,
        'Max CC0': None,
        'Max CC1': None,
        'Max CO': None,
        'Execution Time (ns)': None
    }

    # Read the input file with detected encoding
    with open(file_path, 'r', encoding=encoding) as file:
        lines = file.readlines()

        # Extracting values from specific lines
        for line in lines:
            if line.startswith('No. of Primary Inputs:'):
                parameters['No. of Primary Inputs'] = int(line.split(':')[1].strip())
            elif line.startswith('No. of Primary Outputs:'):
                parameters['No. of Primary Outputs'] = int(line.split(':')[1].strip())
            elif line.startswith('No. of Gates:'):
                parameters['No. of Gates'] = int(line.split(':')[1].strip())
            elif line.startswith('No. of Nodes:'):
                parameters['No. of Nodes'] = int(line.split(':')[1].strip())
            elif line.startswith('Signal'):
                # Skip header lines until the signal table
                break

        # Now extract maximum values of CC0, CC1, and CO
        max_cc0 = float('-inf')
        max_cc1 = float('-inf')
        max_co = float('-inf')
        for line in lines:
            if line.startswith('N'):
                values = extract_numbers(line)
                if len(values) == 4:
                    cc0_value = values[1]
                    cc1_value = values[2]
                    co_value = values[3]
                    if cc0_value > max_cc0:
                        max_cc0 = cc0_value
                    if cc1_value > max_cc1:
                        max_cc1 = cc1_value
                    if co_value > max_co:
                        max_co = co_value
            elif line.startswith('Execution Time:'):
                parameters['Execution Time (ns)'] = extract_execution_time(line)

        parameters['Max CC0'] = max_cc0
        parameters['Max CC1'] = max_cc1
        parameters['Max CO'] = max_co

    return parameters

# Function to process all files in a folder
def process_folder(folder_path, output_excel):
    # Initialize Excel workbook and worksheet
    wb = openpyxl.Workbook()
    ws = wb.active

    # Process each file in the folder
    header_written = False
    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            file_path = os.path.join(folder_path, filename)
            try:
                file_data = process_file(file_path)
                if not header_written:
                    # Write headers
                    headers = list(file_data.keys())
                    ws.append(headers)
                    header_written = True

                # Write data for each file in a new row
                row_data = [file_data[header] for header in headers]
                ws.append(row_data)
            except Exception as e:
                print(f"Error processing {filename}: {str(e)}")

    # Save the Excel workbook
    wb.save(output_excel)
    print(f"Data has been written to {output_excel}")

# Example usage:
if __name__ == "__main__":
    folder_path = '/content/c6288'  # Replace with the path to your folder containing .txt files
    output_excel = '/content/c6288/output.xlsx'  # Output Excel file name

    process_folder(folder_path, output_excel)
```
